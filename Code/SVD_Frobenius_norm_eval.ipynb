```python
# Import packages 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
```

```python
# Import data
data = pd.read_csv('../recipe_ingredients_dataset/ingredient_df_with_recipenames.csv')

# Remove categorical columns 
recipes_ingredients_matrix = data.select_dtypes(include=[np.number])
```

```python
# Convert binary recipe matrix to float
recipes_ingredients_matrix_float = recipes_ingredients_matrix.astype(np.float64)
```

```python
errors = []

# Number of components for Truncated SVD
n_components_test = [100, 500, 1000, 1500, 2000, 3000]

for k in n_components_test:

    # Run Truncated SVD with randomized algorithm for optimal k vectors
    truncated_svd = TruncatedSVD(n_components= k, algorithm='randomized', random_state=42)

    # Fit and transform the data to reduce its dimensions
    X_reduced = truncated_svd.fit_transform(recipes_ingredients_matrix_float)
    
    # Reconstruct the matrix from the reduced dimensions
    X_reconstructed = truncated_svd.inverse_transform(X_reduced)
    
    # Calculate the Frobenius norm of the reconstruction error
    reconstruction_error = np.linalg.norm(recipes_ingredients_matrix_float - X_reconstructed, 'fro')
    
    errors.append(reconstruction_error)
    
```

```python
relative_errors = []

# Compute the original Frobenius norm of the matrix
original_norm = np.sqrt(np.sum(np.square(recipes_ingredients_matrix_float.to_numpy()))) 

for error in errors:

    relative_error = reconstruction_error / original_norm
    
    relative_errors.append(relative_error)

relative_errors
```

```python
# Plotting the reconstruction error
plt.figure(figsize=(10, 6))
plt.plot(n_components_test, errors, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Reconstruction Error (Frobenius Norm)')
plt.title('Reconstruction Error by Different Numbers of Components')
plt.grid(True)
plt.show()
```

# Operationalizing Frobenius Norm in Sparse Data Context

For a binary matrix, the Frobenius norm represents the square root of the sum of the squared differences between the original and reconstructed matrices. Given that the values are either 0 or 1, the Frobenius norm is effectively counting the number of ingredient presences (1's) that are incorrectly reconstructed. 

In this case:

High Precision: A high precision in reconstructing the binary matrix could be important for ensuring the unique identity of each recipe is maintained.

General Patterns: If the recommendation system works well by capturing general patterns rather than exact matches, a higher reconstruction error may be tolerable.


```python
# Plotting the Relative error
plt.figure(figsize=(10, 6))
plt.plot(n_components_test, relative_errors, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Relative Error')
plt.title('Relative Error by Different Numbers of Components')
plt.grid(True)
plt.show()
```
