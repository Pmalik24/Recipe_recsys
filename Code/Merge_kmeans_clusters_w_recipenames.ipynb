```python
#Imports

import pandas as pd 
```

```python
# Load data

data = pd.read_csv('../recipe_ingredients_dataset/ingredient_df_with_recipenames.csv', index_col = 'id',  usecols=['id', 'recipename'])
kmeans = pd.read_csv('../recipe_ingredients_dataset/pca_kmeans_data.csv', index_col = 'id',  usecols=['id', 'Cluster50'])
cuisine = pd.read_csv('../recipe_ingredients_dataset/cluster_cuisine_composition_50.csv', index_col = 0)

cuisine
```

```python
# merge kmeans clusters and recipe names on index 
df_joined = data.join(kmeans)

df_joined 
```

```python
import pandas as pd


# Merge 'df_joined' and 'cuisine' on the 'Cluster50' column
df_final = pd.merge(df_joined, cuisine, on='Cluster50', how='left')

df_final
```

```python
df_final.to_csv('../recipe_ingredients_dataset/pca_clusters.csv')
```
