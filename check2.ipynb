```python
df = pd.read_csv('recipe_ingredients_dataset/ingredient_df.csv')

df.head(1)
```

```python
def clean_column_names(column):
    # Convert to lowercase
    column = column.lower()
    # Replace or remove specific characters
    column = column.replace('®', '')
    # Strip any potential leading/trailing whitespace that may have been added accidentally
    column = column.strip()
    return column

# Apply the cleaning function to each column name
df.columns = [clean_column_names(col) for col in df.columns]
```
