The Below Function takes in an input, and matches it to what is likely the most close thing to it. we will use to process input strings, so that we can output a matrix and map back etc. 

need a function that will take an input string, process it just like parth did the first ingredient names, and the fuzzy word match and the closest match returns a 1, 0. 

```python
from rapidfuzz import process, fuzz

def get_closest_match_rapid(query, choices, scorer=fuzz.WRatio, limit=1):
    """
    Returns the closest match(es) of `query` in `choices` using Rapidfuzz.
    - `scorer` can be any of the scoring functions available from fuzz, such as fuzz.WRatio, fuzz.QRatio, etc.
    - `limit` determines the number of top matches to return.
    """
    results = process.extract(query, choices, scorer=scorer, limit=limit)
    return results
```

```python
# Create a Random List to test our recommender

from random import shuffle

def create_binary_list(A, B):
    if A > B:
        raise ValueError("A cannot be greater than B")

    # Create a list with A ones and (B-A) zeros
    binary_list = [1] * A + [0] * (B - A)
    
    # Shuffle the list to randomize the positions of ones and zeros
    shuffle(binary_list)
    
    return binary_list

samplerecipe = create_binary_list(10, 7137)
```

```python
# Make sample a Series for manipulation
x = pd.Series(samplerecipe)
```

```python
# Dot product of our sample and the U matrix. 
xcon = x.dot(vmatrix)
```

```python
xcon
```

```python
# xconcept and U must both be numpy arrays, and must be properly reshaped. 

from sklearn.metrics.pairwise import cosine_similarity

xcon = np.array(xcon)

umatrix = np.array(umatrix)

xcon = xcon.reshape(1, -1)
```

```python
umatrix
```

```python
#cosine similarity with our U matrix
similarities = cosine_similarity(umatrix, xcon)
```

```python
similarities 
```

```python
#Get the Top n scores from my cosine similarity matrix

def top_n_scores(scores, n):
    # Flatten the array (if it's multidimensional)
    scores = scores.flatten()
    
    # Get the indices that would sort the array, and select the last 'n' indices (top scores)
    top_indices = np.argsort(scores)[-n:][::-1]
    
    # Extract the top 'n' scores using these indices
    top_scores = scores[top_indices]
    
    return top_scores, top_indices

top_scores, top_indices = top_n_scores(similarities, 10)
print("Top Scores:", top_scores)
print("Indices of Top Scores:", top_indices)
```

```python
# See what Recipes were returned. 

for each in top_indices:
    print(data.recipename[data.id == each])
```

```python
# Returns the Ingredients in whatever recipe was recommended. 

def get_columns_with_ones(df, row_index):
    condition = df.loc[row_index] == 1
    return df.columns[condition].tolist()

for each in top_indices:
    print(get_columns_with_ones(data,each))
```
